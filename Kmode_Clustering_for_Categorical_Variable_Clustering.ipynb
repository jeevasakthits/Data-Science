**Kmode Model Creation and prediction**

```

km = KModes(n_clusters=23, init='Huang', n_init=5, verbose=1)

km = km.fit(data1)


clusters = km.predict(data1)
# Print the cluster centroids
print(km.cluster_centroids_)

    
```
