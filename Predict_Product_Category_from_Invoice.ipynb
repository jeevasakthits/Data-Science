```
# Load library
from nltk.corpus import stopwords

# You will have to download the set of stop words the first time
import nltk
nltk.download('stopwords')
```

```
train["Item_Description"]=train["Item_Description"].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```
stop = stopwords.words('english')
```

```
train["Item_Description"]=train["Item_Description"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```
train["Item_Description"]=train["Item_Description"].str.replace('\d+', '')

train["Item_Description"]=train["Item_Description"].str.replace('\([^()]*\)', '')

train["Item_Description"]=train["Item_Description"].str.replace('[!@#$%^&*(),.?":{}|<>]', '')

train["Item_Description"]=train["Item_Description"].str.replace('[+]', '')

train["Item_Description"]=train["Item_Description"].str.replace('[-]', '')

train["Item_Description"]=train["Item_Description"].str.replace('[/]', ' ')
```

```

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd


tfidf=TfidfVectorizer(min_df=2,max_df=0.5,ngram_range=(1,2))


train['Item_Description']=train['Item_Description'].astype(str)

features=tfidf.fit(train['Item_Description'])

measures=features.transform(train['Item_Description'])


df=pd.DataFrame(
        measures.todense(),
        columns=tfidf.get_feature_names()
        )


df.head()

df.columns

```
