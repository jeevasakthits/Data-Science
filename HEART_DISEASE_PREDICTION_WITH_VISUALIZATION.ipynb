````r

###MODEL

#build logistic regression model
glm_mod = glm(as.factor(target) ~ Age + sex + exang + oldpeak + thalach + slope + ca, data=train, family = "binomial")
glm_mod
glm_pred = predict(glm_mod, newdata = train, type = "response")
glm_pred
````

````r

#ROC curve
ROCRpred = prediction(glm_pred, train$target)
plot(performance(ROCRpred,"tpr","fpr"), colorize=TRUE)
auc_glm=round(as.numeric(performance(ROCRpred, "auc")@y.values), digits = 3)
````

````r

#AUC and model accuracy
table(train$target, glm_pred > .35)
model_acc = round((70 + 114)/nrow(train), digits = 3)
print(paste("AUC for Logistic Regression is: ", auc_glm))
print(paste("model accuracy Logistic Regression with a .35 threshold is :", model_acc))
````

````r

#fit a Naive Bayes model
bay_mod = naive_bayes(as.factor(target) ~ Age + sex + exang + oldpeak + thalach + slope + ca, data=train)
bay_mod
bay_pred = predict(bay_mod, newdata=train, type = "prob")
bay_pred
bay_pred = round(bay_pred[,2], digits = 5)
bay_pred
````

````r

#ROC curve
ROCRpred = prediction(bay_pred, train$target)
plot(performance(ROCRpred,"tpr","fpr"), colorize=TRUE)
auc_bay=round(as.numeric(performance(ROCRpred, "auc")@y.values), digits = 3)

````

````r

#AUC and model accuracy
table(train$target, bay_pred > .35)
model_acc = round((78+110)/nrow(train), digits = 3)
print(paste("AUC for Logistic Regression is: ", auc_bay))
print(paste("model accuracy Logistic Regression with a .35 threshold is :", model_acc))

````

````r

#fit a Decision Tree Model
tree_mod = rpart(as.factor(target) ~ Age + sex + exang + oldpeak + thalach + slope + ca, data=train, method="class")
tree_mod
pred_tree = predict(tree_mod, newdata = train)
pred_tree
pred_tree = round(pred_tree[,2], digits=5)
pred_tree
````

````r
#fit ROC curve
ROCRpred = prediction(pred_tree, train$target)
plot(performance(ROCRpred, "tpr", "fpr"), colorize = TRUE)
````

````r

#auc and model accuracy
table(train$target, pred_tree > .35)
model_acc = round((75+116)/nrow(train), digits=3)
auc_tree = round(as.numeric(performance(ROCRpred, "auc")@y.values), digits=3)
print(paste("Decision Tree model AUC is:", auc_tree))
print(paste("Decision Tree model accuracy as .35 threshold is:", model_acc))
````
